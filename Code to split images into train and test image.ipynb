# Code to split images into train and test image

```python
import os
import csv
import shutil
from os import listdir
```

## Train images

```python
# Upload training images name from "train.csv"
train_image_name = []
with open('./dataset/dataset/train.csv','r')as file:
    filecontent= csv.reader(file)
    for row in filecontent:
        train_image_name.append(row[1])
```

```python
# a single image contains multiple labels, so we remove the dublicate image names
revised_train_image_name = [*set(training_image_name[1::])]
```

```python
# copy images with name in "revised_training_image_name" list to a folder train
for i in range(len(revised_training_image_name)):
    for images in os.listdir("./dataset/dataset/images"):
        if images == revised_training_image_name[i]:
            shutil.copy("./dataset/dataset/images/"+str(revised_training_image_name[i]), "./dataset/dataset/images/train/"+str(revised_training_image_name[i]))
            print(images)
```

## Test images

```python
# Upload training images name from "test.csv"
test_image_name = []
with open('./dataset/dataset/test.csv','r')as file:
    filecontent= csv.reader(file)
    for row in filecontent:
        test_image_name.append(row[0])
```
